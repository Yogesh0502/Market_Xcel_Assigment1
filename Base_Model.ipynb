```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pickle
```

```python
df=pd.read_csv('Assignment 1_sales data.csv')
```

```python
'''
From a lot of research, I saw that for RAM and Internal moeory missing,
the price are low compare to others.
So, I am imputing a different Category for those having missing RAM and Internal Memory.
'''
```

```python
df.RAM=df.RAM.fillna('Feature1_0')
df['Internal Memory']=df['Internal Memory'].fillna('Feature2_0')
```

```python
df.head()
```

```python
'''
Since the problem statment is to
Create a webapp where the user can input city, brand, model name, RAM and internal memory to predict Quantity and Amount.
First we will try to predict the Amount. So, we are not considering the Quantity for now. 
As, there are only 10 data points for Quantity > 1. We will deal with them later
'''
```

```python
df_Q_1=df[df.Quantity==1]
```

```python
df_Q_1.info()
```

```python
#Now since the app should contain city, brand, model name, RAM and internal memory. Droping all the other features
```

```python
df_Q_1.drop(['serialNo','Quantity','Date'],inplace=True,axis=1)
```

```python
df_Q_1.head()
```

```python
X=df_Q_1.drop(['Amount'],axis=1)
```

```python
y=df_Q_1.loc[:,['Amount']]
```

```python
X.head(50)
```

```python
y.head()
```

```python
#Lets do some Label Encoding first
```

```python

le_1=LabelEncoder()
le_2=LabelEncoder()
le_3=LabelEncoder()
le_4=LabelEncoder()
le_5=LabelEncoder()
```

```python
X.iloc[:,0]=le_1.fit_transform(X.iloc[:,0])
X.iloc[:,1]=le_2.fit_transform(X.iloc[:,1])
X.iloc[:,2]=le_3.fit_transform(X.iloc[:,2])
X.iloc[:,4]=le_4.fit_transform(X.iloc[:,4])
X.iloc[:,5]=le_5.fit_transform(X.iloc[:,5])
```

```python
X.head(30)
```

```python
# for simplicity and since model does not effect much, so droping the model column
X.drop(['Model'],inplace=True,axis=1)
X.drop(['Price/Unit'],inplace=True,axis=1)
```

```python
X.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer([('XL',OneHotEncoder(drop='first'),[0,1,2,3])],remainder='passthrough')
```

```python
X=ct.fit_transform(X)
```

```python
cl_name=ct.get_feature_names()
```

```python
cl_name
```

```python
X=pd.DataFrame.sparse.from_spmatrix(X,columns=cl_name)
```

```python
X.head()
```

```python

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg1=LinearRegression()
reg1.fit(X_train,y_train)
```

```python
y_pred1=reg1.predict(X_test)
```

```python

r2=r2_score(y_test,y_pred1)
```

```python
r2
```

```python
from sklearn.ensemble import RandomForestRegressor
reg2=RandomForestRegressor(random_state=0)
reg2.fit(X_train,y_train)
y_pred2=reg2.predict(X_test)
r2_score(y_test,y_pred2)
```

```python
# Model will have input 
pickle.dump(reg2, open('model.pkl','wb'))
pickle.dump(ct,open('column_transformer.pkl','wb'))
```

### Feature Selection

```python
# ANOVA feature selection for categorical input and Numerical output
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
# define feature selection
fs = SelectKBest(score_func=f_regression, k=20)
# apply feature selection
selector = fs.fit(X, y)
cols = selector.get_support(indices=True)
features_df_new = X.iloc[:,cols]
```

```python
X.head()
```

```python
features_df_new.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(features_df_new,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg2=RandomForestRegressor(random_state=0)
reg2.fit(X_train,y_train)
y_pred2=reg2.predict(X_test)
r2_score(y_test,y_pred2)
```

##### So, 20 feature are giving result close to 52 features. Hence, Selecting only 20 features for the model.

## Hyper parameter tunning

```python
reg2.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110,200],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
# Create a based model
rf = RandomForestRegressor()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
reg2=RandomForestRegressor(n_estimators=100,bootstrap=True,max_depth=110,max_features=3,min_samples_leaf=3,min_samples_split=12)
reg2.fit(X_train,y_train)
y_pred2=reg2.predict(X_test)
r2_score(y_test,y_pred2)
```
